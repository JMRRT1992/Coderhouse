Data Set 1: Base de datos de IMDb (películas) con títulos de 1961 a 2015.

Se carga el Data Set.

```python
import pandas as pd
import numpy as np

data_set_1 = pd.read_csv("/workspaces/Coderhouse/Proyecto/Data/Raw/TMBD Movie Dataset.csv")
data_frame_1 = pd.DataFrame(data_set_1)
data_frame_1.head(5)
```

Este Data Set presenta una conjunto de películas que se encuentran en el portal IMDb (www.imdb.com).
El Data Set cuenta con 24 columnas y 10760 títulos desde 1961 hasta 2015.

Las columnas representan las variables de estudio y son:

*ID: número identificador para cada película (arbitrario).

*Popularity: calificación de popularidad.

*Budget: presupuesto ejercido en la realización de la película.

*Revenue: ingresos obtenidos de la película.

*Cast: reparto.

*Homepage: página web de la película (si existió).

*Director: director(es).

*Tagline: slogan publicitario.

*keywords: palabras asociadas con el título.

*Overview: sinápsis de la película.

*Runtime: duración en minutos.

*Genres: categorías.

*Production companies: compañías productoras.

*Release date: fecha de estreno.

*Vote count: cuantos votos se tienen para calificar este título.

*Vote average: promedio de la calificación de cada voto.

*Release year: año de estreno (columna redundante).

*Budget adj: presupuesto ajustado a USD 2010 considerando inflación.

*Revenue adj: ingresos ajustados a USD 2010 considerando inflación.
